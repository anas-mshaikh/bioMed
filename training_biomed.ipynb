```python
import os
import shutil

folder_to_zip = "/content/biomed/outputs/training"
output_zip_file = "/content/biomed_outputs_training.zip"

# Remove existing zip file if it exists
if os.path.exists(output_zip_file):
    os.remove(output_zip_file)

# Create a zip archive of the folder
shutil.make_archive(os.path.splitext(output_zip_file)[0], 'zip', folder_to_zip)

print(f"Folder '{folder_to_zip}' has been zipped to '{output_zip_file}'")
```

After running the cell above, you can find the `biomed_outputs_training.zip` file in the file browser on the left (the folder icon). Right-click on it and select 'Download' to save it to your local machine.
